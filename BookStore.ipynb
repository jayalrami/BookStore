# Add Book to Inventory

```python

class Bookstore:
    def __init__(self, inventory_file="inventory_10000.csv", sales_file="sales_10000.csv"):
        self.inventory_file = inventory_file
        self.sales_file = sales_file
        self.inventory = None
        self.sales = None
    # ------------- inventory DataSet ---------------
    def load_data(self):
        try:
            self.inventory = pd.read_csv(self.inventory_file)
            print(" Inventory loaded!")
        except FileNotFoundError:
            print(" Inventory file not found, creating new one...")
            self.inventory = pd.DataFrame(columns=["Title", "Author", "Genre", "Price", "Quantity"])
         # ------------- sales DataSet ---------------
        try:
            self.sales = pd.read_csv(self.sales_file)
            print(" Sales loaded!")
        except FileNotFoundError:
            print(" Sales file not found, creating new one...")
            self.sales = pd.DataFrame(columns=["Date", "Title", "Quantity Sold", "Total Revenue"])

    # ================= ADD BOOK =================
    def add_book(self, title, author, price, quantity):
        if title in self.inventory:
            print(f"Book '{title}' already exists. Updating quantity instead.")
            self.inventory[title]["quantity"] += quantity
        else:
            self.inventory[title] = {
                "author": author,
                "price": price,
                "quantity": quantity
            }
        print(f"Book '{title}' added successfully!")
         
            
store = Bookstore()
store.load_data()
store.add_book('j.k.shah',8000,9,4)
```

# Record sale

```python

class Bookstore:
    def __init__(self, inventory_file="inventory_10000.csv", sales_file="sales_10000.csv"):
        self.inventory_file = inventory_file
        self.sales_file = sales_file
        self.inventory = None
        self.sales = None
    # ------------- inventory DataSet ---------------
    def load_data(self):
        try:
            self.inventory = pd.read_csv(self.inventory_file)
            print(" Inventory loaded!")
        except FileNotFoundError:
            print(" Inventory file not found, creating new one...")
            self.inventory = pd.DataFrame(columns=["Title", "Author", "Genre", "Price", "Quantity"])
         # ------------- sales DataSet ---------------
        try:
            self.sales = pd.read_csv(self.sales_file)
            print(" Sales loaded!")
        except FileNotFoundError:
            print(" Sales file not found, creating new one...")
            self.sales = pd.DataFrame(columns=["Date", "Title", "Quantity Sold", "Total Revenue"])
             
    # ================= Record sale  =================
    def record_sale(self, title, quantity):
        if title not in self.inventory:
            print("Book not found in inventory!")
            return
         
        if self.inventory[title] < quantity:
            print("Not enough stock available!")
            return
            
        self.inventory[title] -= quantity
        
        # Update sales data
        if title in self.sales:
            self.sales[title] += quantity
        else:
            self.sales[title] = quantity
        
        print(f"Sale recorded: {title} -> {quantity} books sold")
    
        
store = Bookstore()
store.load_data()
store.record_sale('j.k.shah',9)            
```

# generate_report

```python
import pandas as pd

class Bookstore:
    def __init__(self, inventory_file="inventory_10000.csv", sales_file="sales_10000.csv"):
        self.inventory_file = inventory_file
        self.sales_file = sales_file
        self.inventory = None
        self.sales = None

    # -------- Load Data --------
    def load_data(self):
        try:
            self.inventory = pd.read_csv(self.inventory_file)
            print(" Inventory loaded!")
        except FileNotFoundError:
            print(" Inventory file not found, creating new one...")
            self.inventory = pd.DataFrame(columns=["Title", "Author", "Genre", "Price", "Quantity"])

        try:
            self.sales = pd.read_csv(self.sales_file)
            print("Sales loaded!")
        except FileNotFoundError:
            print(" Sales file not found, creating new one...")
    # ------------ generate data ---------------        
    def generate_report(self):
        print("\n INVENTORY REPORT")
        for i in range(len(self.inventory)):
            title = self.inventory.loc[i, "Title"]
            qty = self.inventory.loc[i, "Quantity"]
            print(title, ":", qty, "books left")

        print("\n SALES REPORT")
        total = 0
        for i in range(len(self.sales)):
            title = self.sales.loc[i, "Title"]
            sold = self.sales.loc[i, "Quantity Sold"]
            print(title, ":", sold, "books sold")
            total += sold

        print("\nTotal Books Sold:", total)


store = Bookstore()
store.load_data()
store.generate_report()
```

# Numpy and pandas  Operation

```python
import numpy as np

class Bookstore:
    def __init__(self, inventory_file="inventory_10000.csv", sales_file="sales_10000.csv"):
        self.inventory_file = inventory_file
        self.sales_file = sales_file
        self.inventory = None
        self.sales = None

    # ----------- Load Inventory & Sales Data -----------
    def load_data(self):
        try:
            self.inventory = pd.read_csv(self.inventory_file)
            print(" Inventory loaded!")
        except FileNotFoundError:
            print(" Inventory file not found, creating new one...")
            self.inventory = pd.DataFrame(columns=["Title", "Author", "Genre", "Price", "Quantity"])

        
        try:
            self.sales = pd.read_csv(self.sales_file)
            print(" Sales loaded!")
        except FileNotFoundError:
            print(" Sales file not found, creating new one...")
            self.sales = pd.DataFrame(columns=["Date", "Title", "Quantity Sold", "Total Revenue"])

    # ----------- NumPy Revenue Calculation -----------
    def compute_revenue(self):
        if self.inventory is None or self.inventory.empty:
            print(" No inventory data available!")
            return

       
        self.inventory["Revenue"] = self.inventory["Quantity"] * self.inventory["Price"]

        total_revenue = np.sum(self.inventory["Revenue"])
        avg_price = np.mean(self.inventory["Price"])

        print("\n Revenue Analysis")
        print("Total Revenue:", total_revenue)
        print("Average Price:", avg_price)

    # ----------- Best Selling Books -----------
    def best_selling_books(self):
        if self.inventory is None or self.inventory.empty:
            print(" No inventory data available!")
            return
        
        best_books = self.inventory.sort_values(by="Quantity", ascending=False)
        print("\n Best Selling Books:")
        print(best_books[["Title", "Quantity"]].head())


store = Bookstore()
store.load_data()
store.compute_revenue()
store.best_selling_books()


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


class Bookstore:
    def __init__(self, inventory_file="inventory_10000.csv", sales_file="sales_10000.csv"):
        self.inventory_file = inventory_file
        self.sales_file = sales_file
        self.inventory = None
        self.sales = None

    # ----------- Load Inventory & Sales Data -----------
    def load_data(self):
        try:
            self.inventory = pd.read_csv(self.inventory_file)
            print("Inventory loaded!")
        except FileNotFoundError:
            print("Inventory file not found, creating new one...")
            self.inventory = pd.DataFrame(columns=["Title", "Author", "Genre", "Price", "Quantity"])

        try:
            self.sales = pd.read_csv(self.sales_file)
            print("Sales loaded!")
        except FileNotFoundError:
            print("Sales file not found, creating new one...")
            self.sales = pd.DataFrame(columns=["Date", "Title", "Quantity Sold", "Total Revenue"])

    # ----------- BAR CHART: Total Sales by Genre -----------
    def Barchart(self):
        merged = pd.merge(self.sales, self.inventory, on="Title", how="left")
        genre_sales = merged.groupby("Genre")["Total Revenue"].sum().sort_values(ascending=False)
        plt.figure(figsize=(10, 6))
        genre_sales.plot(kind="bar")
        plt.title("Total Sales by Genre")
        plt.xlabel("Genre")
        plt.ylabel("Total Revenue")
        plt.xticks(rotation=45)
        plt.show()

store = Bookstore()
store.load_data()
store.Barchart()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class Bookstore:
    def __init__(self, inventory_file="inventory_10000.csv", sales_file="sales_10000.csv"):
        self.inventory_file = inventory_file
        self.sales_file = sales_file
        self.inventory = None
        self.sales = None

    def load_data(self):
        try:
            self.inventory = pd.read_csv(self.inventory_file)
            print("Inventory loaded!")
        except FileNotFoundError:
            print("Inventory file not found, creating new one...")
            self.inventory = pd.DataFrame(columns=["Title", "Author", "Genre", "Price", "Quantity"])

        try:
            self.sales = pd.read_csv(self.sales_file)
            print("Sales loaded!")
        except FileNotFoundError:
            print("Sales file not found, creating new one...")
            self.sales = pd.DataFrame(columns=["Date", "Title", "Quantity Sold", "Total Revenue"])

    def heatmap(self):
        merged = pd.merge(self.sales, self.inventory, on="Title", how="left")
        corr_data = merged[["Price", "Quantity Sold", "Total Revenue"]].corr()
        plt.figure(figsize=(6, 5))
        sns.heatmap(corr_data, annot=True, cmap="coolwarm")
        plt.title("Correlation Heatmap")
        plt.show()

store = Bookstore()
store.load_data()
store.heatmap()
```
